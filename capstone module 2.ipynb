# Deskripsi Dataset

Dataset ini berisi data transaksi penjualan produk SaaS (Software-as-a-Service) oleh sebuah perusahaan. Setiap baris mewakili satu transaksi individual yang telah dilakukan oleh pelanggan dari berbagai negara dan industri. Data ini berguna untuk menganalisis performa penjualan berdasarkan wilayah geografis, jenis pelanggan, dan produk, serta mengevaluasi dampak diskon terhadap profitabilitas.

Berikut adalah penjelasan masing-masing kolom dalam dataset ini:

- Row ID: ID unik untuk setiap transaksi

- Order ID: ID unik untuk setiap pesanan

- Order Date: Tanggal saat pesanan dilakukan

- Date Key: Representasi numerik dari tanggal pesanan dalam format YYYYMMDD

- Contact Name: Nama orang yang melakukan pesanan

- Country: Negara tempat pesanan dilakukan

- City: Kota tempat pesanan dilakukan

- Region: Wilayah geografis tempat pesanan dilakukan

- Subregion: Sub-wilayah tempat pesanan dilakukan

- Customer: Nama perusahaan atau organisasi yang melakukan pesanan

- Customer ID: ID unik untuk setiap pelanggan

- Industry: Industri tempat pelanggan berasal

- Segment: Segmentasi pelanggan (misalnya: SMB, Strategic, Enterprise, dll.)

- Product: Produk yang dipesan

- License: Kunci lisensi untuk produk yang dibeli

- Sales: Total nilai penjualan untuk transaksi tersebut (dalam mata uang USD)

- Quantity: Jumlah unit produk yang dibeli dalam transaksi tersebut

- Discount: Diskon yang diberikan untuk transaksi tersebut (dalam persentase desimal, misalnya 0.2 untuk 20%)

- Profit: Keuntungan bersih yang diperoleh dari transaksi tersebut setelah dikurangi biaya

Grafik memperlihatkan bahwa distribusi profit sangat tidak simetris dan tidak mengikuti distribusi normal. Puncak distribusi sangat tinggi dan sempit di sekitar titik nol, sementara sisa distribusinya memiliki ekor yang panjang ke kiri (kerugian) dan ke kanan (keuntungan besar).

Sebagian besar transaksi menghasilkan profit yang sangat kecil atau bahkan nol.

Terdapat skewness positif ringan, karena ekor kanan (profit besar) lebih panjang dari ekor kiri.

Adanya outlier signifikan di kedua sisi, terutama di nilai profit negatif (kerugian besar) dan profit tinggi

Distribusi memperlihatkan bahwa mayoritas transaksi menghasilkan profit yang rendah atau netral. Ini bisa berarti:

Banyak penjualan dilakukan dengan margin tipis.

Diskon yang besar atau biaya tinggi bisa menyebabkan profit menjadi rendah meskipun penjualan (sales) tinggi.

# 7. Tangani Outlier (jika diperlukan)

Jika distribusi Profit menunjukkan adanya outlier ekstrem:

```python
numerical_cols = ['Sales', 'Quantity', 'Discount', 'Profit']
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1

# Tandai data sebagai outlier jika berada di luar rentang IQR
df['Is_Outlier'] = (df['Profit'] < (Q1 - 1.5 * IQR)) | (df['Profit'] > (Q3 + 1.5 * IQR))

# Opsional: Lihat berapa banyak outlier
print(f"Jumlah outlier pada kolom Profit: {df['Is_Outlier'].sum()}")
```

Dari analisa di atas, saya tetap Gunakan semua data, termasuk outlier, karena:

- Outlier merefleksikan kejadian nyata (misalnya penjualan besar),

- Anda ingin menganalisis tren secara menyeluruh.

# 8. Pastikan Konsistensi Format Tanggal

```python
# Filter data hanya untuk region APJ
apj_df = df[df['Region'] == 'APJ'].copy()

# Tambahkan kolom margin jika belum ada
apj_df['Margin'] = apj_df['Profit'] / apj_df['Sales']

# 1. Cek rata-rata diskon di APJ
avg_discount_apj = apj_df['Discount'].mean()

# 2. Analisis produk dengan margin negatif
product_margin = apj_df.groupby('Product')[['Profit', 'Sales']].sum()
product_margin['Margin'] = product_margin['Profit'] / product_margin['Sales']
low_margin_products = product_margin.sort_values('Margin').head(10)

# 3. Analisis pelanggan merugi
customer_margin = apj_df.groupby('Customer')[['Profit', 'Sales']].sum()
customer_margin['Margin'] = customer_margin['Profit'] / customer_margin['Sales']
unprofitable_customers = customer_margin[customer_margin['Margin'] < 0].sort_values('Margin')

# 4. Cek rata-rata margin per segmen pelanggan
segment_margin = apj_df.groupby('Segment')[['Profit', 'Sales']].sum()
segment_margin['Margin'] = segment_margin['Profit'] / segment_margin['Sales']

# Tampilkan hasil
print("Rata-rata Diskon di APJ:", avg_discount_apj)
print("\nTop Produk dengan Margin Terendah di APJ:\n", low_margin_products)
print("\nPelanggan Tidak Menguntungkan di APJ:\n", unprofitable_customers.head(10))
print("\nMargin Keuntungan per Segment Pelanggan di APJ:\n", segment_margin)
```
